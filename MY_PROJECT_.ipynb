FOOD DELIVERY SALES AND CUSTOMER INSIGHTS

```python
from google.colab import files
uploaded=files.upload()
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Food_Delivery1.csv")
df.columns=df.columns.str.strip()
df["SALES"]=df["QUANTITY"]* df["PRICE"]
df["ORDER_TIME"] = df["ORDER_TIME"].str.strip()
df["ORDER_TIME"]=pd.to_datetime(df["ORDER_TIME"],format="%H:%M")
df["HOUR"]=df["ORDER_TIME"].dt.hour
print("FOOD DELIVERY DATA")
print(df)
print("\nTotal Sales=",df["SALES"].sum())
peak=df.groupby("HOUR")["ORDER_ID"].count()
print("\nPeak Order Time")
print(peak)

popular=df.groupby("FOOD_ITEM")["QUANTITY"].sum()
print("\n Most popular food items")
print(popular)
location=df.groupby("LOCATION")["SALES"].sum()
print("\nSales by Location")
print("location")
print("\nOrders from Coimbatore")
print(df[df["LOCATION"]=="Coimbatore"])
print("\nPizza Orders")
print(df[df["FOOD_ITEM"]=="Pizza"])
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
peak.plot(kind="bar")
plt.title("Peak Order Time")
plt.xlabel("Hour")
plt.ylabel("Orders")
plt.subplot(1,3,2)
popular.plot(kind="bar")
plt.title("Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Quantity")
plt.subplot(1,3,3)
location.plot(kind="bar")
plt.title("Sales by Location")
plt.xlabel("Location")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```
